```python
def check_restock_alert(inventory_list):
    """
    Identifies items requiring a restock alert based on low stock and high demand.

    Args:
        inventory_list (list): A list of item dictionaries.

    Returns:
        list: A list of item names that need a restock alert.
    """
    restock_alerts = [] # This list will hold the names of the flagged items

    for item in inventory_list:
        # 1. Extract necessary variables for clarity
        quantity = item["stock"]
        demand = item["sales"]
        status = item["status"]

        # 2. Apply ALL restock alert conditions (must be True for the alert to trigger)
        # Condition 1: Low Quantity (stock < 15)
        # Condition 2: High Demand (sales >= 10)
        # Condition 3: No Backorder (status != "Backordered")
        is_alert = (quantity < 15) and (demand >= 10) and (status != "Backordered")

        # 3. Conditional Appending
        if is_alert:
            restock_alerts.append(item["name"])

    return restock_alerts

# --- Testing the function with the provided data ---
inventory_data = [
    {"name": "Widget A", "stock": 10, "sales": 20, "status": "In Stock"},
    {"name": "Gadget B", "stock": 50, "sales": 5, "status": "In Stock"},
    {"name": "Thing C", "stock": 12, "sales": 15, "status": "Backordered"},
    {"name": "Doodad D", "stock": 5, "sales": 18, "status": "In Stock"},
]

print(check_restock_alert(inventory_data))
```

```python
class Thermostat:
    # 1. Class Attributes (MIN/MAX_TEMP)
    MIN_TEMP = 10
    MAX_TEMP = 30

    def __init__(self):
        # 2. Initialize private instance attribute
        # Starts the target temperature at 20 degrees.
        self._target_temp = 20

    # 3. Getter method
    def get_target_temp(self):
        # Returns the current private target temperature.
        return self._target_temp

    # 4. Setter method with safety logic
    def set_target_temp(self, new_temp):
        # The safety check uses an 'or' operator:
        # If the new_temp is too low OR too high, the 'if' condition is True.
        if new_temp < self.MIN_TEMP or new_temp > self.MAX_TEMP:
            print(f"Error: Temperature {new_temp} out of safe bounds. Must be between {self.MIN_TEMP}Â°C and {self.MAX_TEMP}Â°C.")
        else:
            # If the temperature is safe, update the private attribute.
            self._target_temp = new_temp
            print(f"Target set to {new_temp}Â°C.")
```

```python
def find_two_sum(number_list, target_sum):
    numbers_seen = {}
    for index, current_num in enumerate(number_list):
        complement_needed = target_sum - current_num
        if complement_needed in numbers_seen:
            complement_index = numbers_seen[complement_needed]
            current_index = index
            return [complement_index, current_index]
        numbers_seen[current_num] = index
    return []
numbers = [3, 5, 2, 8]
target = 10 

result = find_two_sum(numbers, target)
print(f"The indices are: {result}")
```

```python
# --- Core Logic Block ---

def find_pair_indices(full_list, desired_total):
    
    # This dictionary is our "Memory Bank" or "Scratchpad."
    # We store numbers we've already checked here for fast lookups.
    # Key = The number we've seen.
    # Value = The index (position) where we saw it.
    memory_of_numbers = {}

    # Iterate through the list, getting both the position and the number itself.
    for current_index, current_number in enumerate(full_list):

        # 1. LOGIC STEP: Determine the required partner
        # This is the specific number we NEED to find to hit the desired_total.
        partner_needed = desired_total - current_number

        # 2. CHECK THE MEMORY: Have we seen the partner_needed before?
        if partner_needed in memory_of_numbers:
            
            # SUCCESS! The partner is in the dictionary.
            
            # The partner's index is stored as the value in our dictionary.
            partner_index = memory_of_numbers[partner_needed]
            
            # The current number's index is 'current_index'.
            
            # Return both indices as the final answer.
            return [partner_index, current_index]

        # 3. UPDATE MEMORY: If no pair was found, add the current number to the bank
        # We store the current number as the key and its position as the value.
        memory_of_numbers[current_number] = current_index

    # If the loop finishes without finding a pair, return an empty list.
    return []

# --- Example Usage ---
my_list = [1, 9, 3, 7]
target_value = 10
result = find_pair_indices(my_list, target_value)
print(result) # Output: [1, 3] because 9 (index 1) + 7 (index 3) = 10
```

```python
class Solution:
       def twoSum(self, nums: list[int], target: int) -> list[int]:
          prevMap = {} # Initialize a dictionary (hash map) to store (value, index)
          for i, n in enumerate(nums): # Loop through the list with index i and value n
             diff = target - n # Calculate the required difference
             if diff in prevMap: # Check if the difference needed is already in the map
                return [prevMap[diff], i] # Found the pair: return the stored index and the current index
             prevMap[n] = i # If not found, add the current number and its index to the map
          
          return [] # Only return an empty list if the loop completes without finding a solution

# Example execution
List = [-2, 1, 0, 4, -3]
target = 1
result = Solution().twoSum(List, target)
print(result)
```

```python
# THE PARENT
class Phone:
    def display_message(self):
        print("Message: Battery Low.")

# THE CHILD (Overrides the message)
class SmartPhone(Phone):
    def display_message(self):
        # This replaces the Parent's version entirely
        print("Notification: You have 5 new Instagram likes!")

# --- EXECUTION ---
basic_phone = Phone()
my_iphone = SmartPhone()

basic_phone.display_message() # Output: Message: Battery Low.
my_iphone.display_message()   # Output: Notification: You have 5 new Instagram likes!
```
